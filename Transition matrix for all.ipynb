```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt; plt.rcdefaults()
```

```python
data = pd.read_excel('walks_all.xlsx')
```

```python
del data['Unnamed: 0']
```

```python
data
```

```python
data_np= np.array(data.iloc[:, 1:].values)
data_np= data_np.astype('float64') 

```

```python

v, h = data.index, data.index

fig, ax = plt.subplots()
im = ax.imshow(data_np)

# We want to show all ticks...
ax.set_xticks(np.arange(len(h)))
ax.set_yticks(np.arange(len(v)))

# ... and label them with the respective list entries
ax.set_xticklabels(h)
ax.set_yticklabels(v)

# Rotate the tick labels and set their alignment.
plt.setp(ax.get_xticklabels(), rotation=90, ha="right",
         rotation_mode="anchor")

# Loop over data dimensions and create text annotations.
for i in range(len(v)):
    for j in range(len(h)):

        text = ax.text(j, i, data_np[i, j], ha="center", va="center", color="w")


ax.set_title("Probability for the next app")
fig.tight_layout()

plt.show()
```
