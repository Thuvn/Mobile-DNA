```python
dataLU2 = data_apps.iloc[:,10:15]
```

```python
def densplot(data):
    data.columns = ['1app', '2apps', '3apps', '4apps', '5apps']
    df = data
    p1=sns.kdeplot(df['1app'], shade=True, color="c") 
    p1=sns.kdeplot(df['2apps'], shade=True, color="k")
    p1=sns.kdeplot(df['3apps'], shade=True, color="r") 
    p1=sns.kdeplot(df['4apps'], shade=True, color="y")
    p1=sns.kdeplot(df['5apps'], shade=True, color="grey") 
```

```python
densplot(dataLU)
```

```python
densplot(dataLU2)
```

```python
densplot(dataBM)
```

b: blue
