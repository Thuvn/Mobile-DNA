# Corr between number of apps and accuracy of model

```python
import pandas as pd
```

```python
accu = pd.read_excel('data_apps.xlsx') # LU, LU-2, MFU (called BM)
```

```python
LUT = pd.read_excel('LUT_result.xlsx')
```

```python
LUT.head()
```

```python
info = pd.read_csv('result1.csv')
```

```python
accu.head()
```

```python
info.head()
```

```python
result = pd.merge(info, accu, on='id')
```

```python
result2 = pd.concat([result,LUT], axis=1)
```

```python
result2.columns
```

```python
df = result2.loc[:, ['list_num_app', 'list_num_ss1', 'list_num_ss2',
        'BM1', 'BM2', 'BM3', 'BM4', 'BM5',
       'LU1', 'LU2', 'LU3', 'LU4', 'LU5', 'LU2_1', 'LU2_2', 'LU2_3', 'LU2_4',
       'LU2_5',
                   'LUT_1app', 'LUT_2apps', 'LUT_3apps', 'LUT_4apps',
       'LUT_5apps']]

df.corr()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = df.list_num_app
y = df.LU3
plt.scatter(x, y, color = 'b', alpha=0.5)
plt.xlabel("number of apps")
plt.ylabel("accuracy LU - predicted 3apps")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = df.list_num_app
y = df.LU2_3
plt.scatter(x, y, color = 'k', alpha=0.5)
plt.xlabel("number of apps")
plt.ylabel("accuracy LU-2 predicted 3 apps")
plt.show()
```

# LU accuracy and proposed factor

```python
data_app = pd.read_excel('data_apps.xlsx')
factor = pd.read_excel('factor.xlsx')
data = data_app.set_index('id')
f = factor.set_index('newid')
dataf = data.join(f)
from scipy.stats.stats import pearsonr
pearsonr(dataf.factor, dataf.LU1)[0] #0.743777839
```

```python
import seaborn as sns
sns.set(style="darkgrid")

g = sns.jointplot("factor", "LU1", data=dataf, kind="reg",
                  xlim=(0,1), ylim=(0, 1), color="m", height=7)
```
